# Installation script
### This script installs all required libraries. It may take several (>5) minutes!
### Prior to running this script, please head to the "Settings" menu just above, and "Enable Extension Manager" (make sure this option is ticked, i.e. enabled).
### Now you can run this installation script from the "Run" menu, selecting "Run All Cells".
### Take a look at the circle in the upper-right side of the screen; if it's black that means it's working.
### Once it finishes working, if everything went right the last two lines below should read something like (with a red background; that's OK):

<span style='background:#FFDDDD'> Building jupyterlab assets (build:prod:minimize)  </span>

<span style='background:#FFDDDD'> Building jupyterlab assets (build:prod:minimize)  </span>

### Then close and restart Anaconda and Jupyter Lab, and everything should be ready to use the main program.

```python
import sys # We use this module to execute commands in the host computer
!{sys.executable} -m pip install plotly # This gives us access to plots
!{sys.executable} -m pip install ipywidgets # Widgets for jupyterlab
!{sys.executable} -m pip install ipyfilechooser # This particular widget is used to save files
!{sys.executable} -m pip install -U kaleido # Needed to generate images and pdf files
!{sys.executable} -m pip install PyPDF2 # Used to merge pdf files
!{sys.executable} -m pip install fpdf # Used to create pdf files
!{sys.executable} -m pip install lmfit # Used for fitting

!conda install --yes --prefix {sys.prefix} nodejs # We need Node.js to install and use extensions to jupyterlab
!jupyter labextension install jupyterlab-plotly
!jupyter labextension install @jupyter-widgets/jupyterlab-manager plotlywidget
```
